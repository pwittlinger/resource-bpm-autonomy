```python
import ast
import re
import pandas as pd
```

```python
def read_schedules(file_name):
    raw = open(file_name).read()

    # Extract all [[...]] groups
    lists = re.findall(r'\[\[.*?\]\]', raw)

    all_records = []
    max_length = 0

    for l in lists:
        parsed = ast.literal_eval(l)
        if len(parsed)>max_length:
            max_length = len(parsed)

    for l in lists:
        parsed = ast.literal_eval(l)
        if (len(parsed)<(max_length)):
            continue
        for item in parsed:
            all_records.append({'iteration': item[0], 'objective': item[1]})

    df = pd.DataFrame(all_records)
    return df
    
```

```python
df = read_schedules("best_schedule.txt")
df.describe()
```
